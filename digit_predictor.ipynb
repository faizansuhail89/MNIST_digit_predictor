```python
# Question 2

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import time
import pickle

# Read test data from csv
test_data = pd.read_csv('mnist_test.csv', 
                    delimiter=',', dtype = int)

# Read train data from csv
train_data = pd.read_csv('mnist_train.csv', 
                   delimiter=',', dtype = int)

x_train = train_data.drop(['label'], axis = 1) / 255.
x_test = test_data.drop(['label'], axis = 1) / 255.

y_train = train_data['label']
y_test = test_data['label']

# Initialize timer
start_time = time.time()

# Initialise the classifier
model = MLPClassifier(solver='sgd', activation='relu', batch_size=1, \
                     alpha=1e-4, hidden_layer_sizes=(100,), verbose=True,\
                     max_iter=200, random_state=1)
# Perform training 
model.fit(x_train, y_train) 

print 'Time taken: %f \n' % (time.time() - start_time)

# Save the model
# filename = 'relu_model.sav'
# pickle.dump(model, open(filename, 'wb'))


# load the model from disk
# model = pickle.load(open('relu_model.sav', 'rb'))

y_pred = model.predict(x_test)

print 'Accuracy: %f' % accuracy_score(y_test, y_pred)
confusion_matrix = confusion_matrix(y_test, y_pred)
print confusion_matrix

error_sample1 = x_test.loc[115].values.reshape(1, -1)
error_sample2 = x_test.loc[124].values.reshape(1, -1)
error_sample3 = x_test.loc[247].values.reshape(1, -1)

prob_samp1 = model.predict_proba(error_sample1)
prob_samp2 = model.predict_proba(error_sample2)
prob_samp3 = model.predict_proba(error_sample3)

# Plot the coefficients
fig, axes = plt.subplots(5, 5)
for coef, ax in zip(model.coefs_[0].T, axes.ravel()):
    ax.matshow(coef.reshape(28, 28), cmap=plt.cm.gray)
    ax.set_xticks(())
    ax.set_yticks(())

#plt.savefig('relu_weights.png', dpi=500, bbox_inches ='tight')
plt.show()
```
